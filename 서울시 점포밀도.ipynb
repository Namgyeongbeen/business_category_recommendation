```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 점포밀도 구하기

```python
# 파일 불러오기
df = pd.read_csv('행정동별 면적.csv')#, encoding='cp949')
df
```

```python
# 자치구, 행정동 가나다순 정렬
df_area = df.sort_values(by=['자치구', '행정동'], ascending=True).reset_index().drop('index', axis=1)
df_area
```

```python
# 점포 수 있는 데이터 불러오기
df_store = pd.read_csv('서울시 상권분석서비스(상권-점포)20~22년.csv', encoding='cp949')
df_store
```

```python
# 행정동 별 점포 수 구하기
# 유사업종점포수 = 점포수 + 프랜차이즈점포수

# 20~22년 평균 점포수
df_stores = df_store.groupby(['자치구', '행정동'])['유사업종점포수'].sum().reset_index()
df_stores['20~22년 분기별 평균 점포수'] = df_stores['유사업종점포수']/12


# 자치구, 행정동 가나다순 정렬
df_stores = df_stores.sort_values(by=['자치구', '행정동'], ascending=True)
df_stores
```

```python
df_area
```

```python
# df_area 데이터와 df_stores 데이터에 행정동 수가 다른데, 같게 만들어주자.
# isin() 사용

df_area = df_area[df_area['행정동'].isin(df_store['행정동'])].reset_index().drop('index', axis=1)
df_stores = df_stores[df_stores['행정동'].isin(df_area['행정동'])].reset_index().drop('index', axis=1)

print((df_area.행정동 == df_stores.행정동).sum())
df_area
```

```python
df_stores[df_stores.행정동=='압구정동']
```

```python
df_area[df_area.행정동=='압구정동']
```

```python
# 점포밀도 구하기
# 점포밀도 = 점포 수 / 행정동 면적

density = df_stores['20~22년 분기별 평균 점포수'] / df_area['면적 (k㎡)']

df_density = pd.DataFrame({'자치구':df_area['자치구'], '행정동':df_area['행정동'], '20~22년 평균 점포수':df_stores['20~22년 분기별 평균 점포수'], '면적(k㎡)':df_area['면적 (k㎡)'], '20~22년 평균 점포밀도(#/k㎡)':density})
df_density.to_csv('서울시 점포밀도.csv')
```

```python
df_density[df_density.자치구=='강남구'].sort_values(by='20~22년 평균 점포밀도(#/k㎡)', ascending=False).head(10)
```

# 각 자치구별 Top5

```python
# 강남구만 먼저 시험적으로...

# df_density에서 강남구만 남기자
gangnam = df_density[df_density['자치구'] == '강남구']

sorted_gangnam = gangnam.sort_values(by='20~22년 평균 점포밀도(#/k㎡)', ascending=False)
sorted_values = sorted_gangnam['행정동'].tolist()

print(f'강남구 점포밀도 Top5 : {sorted_values[:]}')
```

```python
# 다른 25개 구 모두 출력하기
for i in range(0,25):
    gu = df_density[df_density['자치구'] == df_density['자치구'].unique()[i]].reset_index().drop('index', axis=1)
    sorted_gu = gu.sort_values(by='20~22년 평균 점포밀도(#/k㎡)', ascending=False)
    sorted_values = sorted_gu['행정동'].tolist()
    print(f"{df_density.자치구.unique()[i]} 점포밀도 Top5 : {sorted_values[:5]}")
    print('-'*100)
```
